```python
df = pd.read_excel("Students_data.xlsx")
df.head()
```

```python
df['Average'] = (df['Maths'] + df['Science'] + df['English']) / 3
```

```python
def grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 50:
        return 'C'
    else:
        return 'F'

df['Grade'] = df['Average'].apply(grade)
```

```python
def result(avg):
    if avg >= 50:
        return 'Pass'
    else:
        return 'Fail'

df['Result'] = df['Average'].apply(result)

df.head()
```

```python
top5 = df.sort_values(by='Average', ascending=False).head(5)
print(top5[['Name','Average','Grade']])
```

```python
subject_avg = df[['Maths','Science','English']].mean()
print(subject_avg)
```

```python
subject_avg.plot(kind='bar')
plt.title("Subject Wise Average Marks")
plt.ylabel("Average Marks")
plt.show()
```

```python
df['Result'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pass vs Fail")
plt.ylabel("")
plt.show()
```
